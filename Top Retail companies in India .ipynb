```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
soup
```

```python
results =soup.find_all("span",{"class":"lwptoc_item_label"})
results
```

```python
results[1].text
```

```python
details = []
for i in results:
    details.append(i.text)
details
```

```python
del details[0]
details
```

```python
name = []
for i in details:
    name.append(i[1:].strip(". "))
del name[0]
name
```

```python
results1 = soup.find_all("li")
results1
```

```python
mk = results1[30:85]
mk
```

```python
records1 = []
for i in mk:
    records1.append(i.text.replace("\xa0","").split(":")[1].strip())
records1
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(l1,columns = ["CompanyName","Revenue in CR","Retail Stores","Employees","Market Cap in CR","ROE","Sales Growth","Promoters Holding","Debt To Equity","Price To BookValue"])
df
```

```python
df.info()
```

## Converting Dataframe objects into Float & Int objects

```python
df['Revenue in CR'] = df['Revenue in CR'].apply(pd.to_numeric)
```

```python
df['Retail Stores'] = df['Retail Stores'].apply(pd.to_numeric)
```

```python
df['Employees'] = df['Employees'].apply(pd.to_numeric)
```

```python
df['Market Cap in CR'] = df['Market Cap in CR'].apply(pd.to_numeric)
```

```python
df['Debt To Equity'] = df['Debt To Equity'].apply(pd.to_numeric)
```

```python
df['Price To BookValue'] = df['Price To BookValue'].apply(pd.to_numeric)
```

```python
df.info()
```

```python
df.describe()
```

## Visualizing Data from the given Dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
sns.distplot(df["Employees"])
```

## Visualizing Revenue in CR against Companies using Bar plot

```python
plt.bar(df['CompanyName'],df['Revenue in CR'], label="Bar 1")
plt.title("CompanyName VS Revenue")
plt.xlabel("CompanyName")
plt.ylabel("Revenue in CR")
plt.xticks(rotation=-90)
plt.legend()
plt.show()
```

## Visualizing Retail stores against Companies using Bar plot

```python
plt.bar(df['CompanyName'],df['Retail Stores'], label="Bar 1",color = "c")
plt.title("CompanyName VS Retail Stores ")
plt.xlabel("CompanyName")
plt.ylabel("Retail Stores")
plt.xticks(rotation=-90)
plt.legend()
plt.show()
```

## Visualizing Mean and Median of Revenue by Distplot

```python
mr =df["Revenue in CR"].mean()
dr =df["Revenue in CR"].median()

sns.distplot(df["Revenue in CR"])
plt.axvline(mr,color="red")
plt.axvline(dr,color="yellow")
plt.grid(True)
```
